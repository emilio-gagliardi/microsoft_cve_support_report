```python
def build_file_mappings(local_base_path, local_html_file, local_day_plot_file, local_weekly_plot_file,
                        remote_base_path, remote_html_path, remote_plots_path):
    """
    Build a dictionary mapping local file paths to their respective remote destinations.

    Parameters:
    - local_base_path: Base path for local files.
    - local_html_file: File name of the local HTML file.
    - local_day_plot_file: File name of the local daily plot image.
    - local_weekly_plot_file: File name of the local weekly plot image.
    - remote_base_path: Base path for remote files.
    - remote_html_path: Subdirectory under remote base for HTML files.
    - remote_plots_path: Subdirectory under remote base for plot images.

    Returns:
    A dictionary with local paths as keys and remote paths as values.
    """
    file_mappings = {}
    
    # Full local paths
    local_html_path = f"{local_base_path}\\html\\{local_html_file}"
    local_day_plot_path = f"{local_base_path}\\plots\\{local_day_plot_file}"
    local_weekly_plot_path = f"{local_base_path}\\plots\\{local_weekly_plot_file}"
    
    # Map the HTML file
    remote_html_full_path = f"{remote_base_path}/{remote_html_path}/{local_html_file}"
    file_mappings[local_html_path] = remote_html_full_path
    
    # Map the daily plot image
    remote_day_plot_full_path = f"{remote_base_path}/{remote_plots_path}/{local_day_plot_file}"
    file_mappings[local_day_plot_path] = remote_day_plot_full_path
    
    # Map the weekly plot image
    remote_weekly_plot_full_path = f"{remote_base_path}/{remote_plots_path}/{local_weekly_plot_file}"
    file_mappings[local_weekly_plot_path] = remote_weekly_plot_full_path

    return file_mappings
```

```python
def split_product_name(row):
    parts = row['product'].split()
    
    try:
        # Find the version part with more generalized conditions
        version_index = next(i for i, part in enumerate(parts) if part[0].isdigit() and (len(part) == 4 or 'H' in part))
        product_part = " ".join(parts[:version_index])
        version_part = parts[version_index]
        architecture_part = " ".join(parts[version_index + 1:])
    except StopIteration:
        version_index = -1
        version_part = ""
        if '(' in row['product'] and ')' in row['product']:
            product_part, architecture_part = row['product'].split(' (')
            architecture_part = architecture_part.rstrip(')')
        else:
            product_part = row['product']
            architecture_part = ""
    
    # Pre-process product_part to remove 'version' and 'Version' before replacing spaces and converting to lowercase
    product_part = product_part.replace(" version", "").replace(" Version", "")
    
    # Trim any trailing spaces that might result in undesired underscores
    product_part = product_part.strip()

    # Replace spaces with underscores and convert to lowercase
    product_part = product_part.lower().replace(" ", "_")
    
    if architecture_part.startswith("for "):
        architecture_part = architecture_part[4:].lower()
    else:
        architecture_part = architecture_part.lower()

    row['product_name'] = product_part
    row['product_version'] = version_part
    row['product_architecture'] = architecture_part
    
    return row


def generate_hash(*args):
    combined_string = ''.join(map(str, args))
    hash_object = hashlib.sha256(combined_string.encode())
    hash_hex = hash_object.hexdigest()
    return hash_hex
    
def add_row_guid(df, columns_to_process, col_name='id'):
    def process_row(row):
        # Concatenate the values of the specified columns
        concatenated_values = ''.join([str(row[col]) for col in columns_to_process])
        # Generate a hash from the concatenated values
        hash_hex = generate_hash(concatenated_values)
        # Convert the hash to a GUID-like format
        guid = str(uuid.UUID(hash_hex[:32]))
        return guid
    
    # Apply the process_row function to each row and assign the result to the 'id' column
    df[col_name] = df.apply(process_row, axis=1)
    return df


def convert_to_tuple(build_number):
    return tuple(map(int, build_number))
    
```

```python
def preprocess_update_guide_product_build_data(data: pd.DataFrame, columns_to_keep: list) -> pd.DataFrame:
    prepped_df = data[columns_to_keep].copy(deep=True)
   
    # Lowercase other columns
    prepped_df.columns = [col.lower() for col in prepped_df.columns]
    prepped_df.columns = prepped_df.columns.str.replace(' ', '_', regex=True)
    # if columns contains 'download_url'
    if 'download_url' in prepped_df.columns:
        prepped_df = prepped_df.rename(columns={'release_date': 'published', 'details': 'cve_id', 'article': 'kb_id', 'download':'package_type', 'download_url':'package_url', 'details_url':'cve_url'})
    else:
        prepped_df = prepped_df.rename(columns={'release_date': 'published', 'details': 'cve_id', 'article': 'kb_id', 'download':'package_type', 'details_url':'cve_url'})
    # Convert 'published' to datetime object
    prepped_df['published'] = pd.to_datetime(prepped_df['published']).dt.strftime('%d-%m-%Y')
    prepped_df = prepped_df[pd.notnull(prepped_df['build_number']) & (prepped_df['build_number'] != '')]
    # Convert 'build_number' to a list of integers
    prepped_df['build_number'] = prepped_df['build_number'].apply(lambda x: [int(n) for n in x.split('.') if n.isdigit()])
    prepped_df= prepped_df.apply(split_product_name, axis=1)
    cols_for_product_build_id = ['product_name', 'product_version', 'product_architecture', 'build_number', 'published']
    prepped_df = add_row_guid(prepped_df, cols_for_product_build_id, 'product_build_id')
    return prepped_df
```

#### Node 7 - merge_product_build_data_with_msrc_docs

```python
def merge_product_build_data_with_msrc_docs(new_product_build_data, existing_docs_to_augment):
    for cve in existing_docs_to_augment:
        cve_id = cve.get('metadata', {}).get('post_id')  # Use .get() for safer access
        
        # Initialize lists and temporary dictionary to store collected information
        product_build_ids = []
        build_number_tuples = set()  # Store build numbers as tuples for easy sorting
        unique_product_strs = set()
        temp_kb_articles_dict = {}  # Temporary dictionary for kb_articles

        # Iterate over DataFrame rows
        for _, row in new_product_build_data.iterrows():
            if row['cve_id'] == cve_id:
                product_build_ids.append(row['product_build_id'])
                # Store build_number as a tuple of integers
                build_number_tuple = tuple(map(int, row['build_number']))  # Assuming build_number is a list
                build_number_tuples.add(build_number_tuple)
                # Ensure "KB" is prepended to kb_id and avoid duplicates
                kb_key = f"kb{row['kb_id']}"
                if kb_key not in temp_kb_articles_dict:
                    temp_kb_articles_dict[kb_key] = row['article_url']
                # Format each product tuple as a string and add to its set for uniqueness
                product_str = f"{row['product_name'].replace('_', ' ')} {row['product_version']} {row['product_architecture']}"
                unique_product_strs.add(product_str)

        # Sort the build_number_tuples
        sorted_build_number_tuples = sorted(build_number_tuples)
        # Convert sorted tuples back to version strings
        build_numbers_lists = [list(bn_tuple) for bn_tuple in sorted_build_number_tuples]
        # Convert the temporary dictionary into the desired list format for kb_articles
        kb_articles_list = [{kb_id: url} for kb_id, url in temp_kb_articles_dict.items()]
        products_str = list(unique_product_strs)

        # Update the cve metadata
        if 'metadata' not in cve:  # Check if metadata key exists
            cve['metadata'] = {}
        cve['metadata']['product_build_ids'] = product_build_ids
        cve['metadata']['build_numbers'] = build_numbers_lists  # Sorted and converted build numbers
        cve['metadata']['kb_articles'] = kb_articles_list
        cve['metadata']['products'] = products_str
        
    return existing_docs_to_augment
```

```python
def load_augmented_msrc_posts_product_build_docs(updated_msrc_docs):
    print(f"processing {len(updated_msrc_docs)} CVEs")
    for cve in updated_msrc_docs:
        # The cve_id is used to identify the documents to update
        cve_id = cve.get('metadata', {}).get('post_id')

        # Retrieve all documents with the matching post_id
        existing_docs = db_collection.find({'metadata.post_id': cve_id})
        
        for existing_doc in existing_docs:
            update_data = {}

            # Handle 'kb_articles' (as previously detailed)

            # Handle 'product_build_ids'
            existing_product_build_ids = set(existing_doc['metadata'].get('product_build_ids', []))
            new_product_build_ids = set(cve['metadata'].get('product_build_ids', []))
            product_build_ids_to_add = list(new_product_build_ids - existing_product_build_ids)
            if product_build_ids_to_add:
                update_data.setdefault('$addToSet', {}).setdefault('metadata.product_build_ids', {}).setdefault('$each', []).extend(product_build_ids_to_add)

            # Handle 'build_numbers' (adjusted approach for handling lists)
            existing_build_numbers = {tuple(bn) for bn in existing_doc['metadata'].get('build_numbers', [])}
            new_build_numbers = {tuple(bn) for bn in cve['metadata'].get('build_numbers', [])}
            build_numbers_to_add_tuples = new_build_numbers - existing_build_numbers
            build_numbers_to_add = [list(bn) for bn in build_numbers_to_add_tuples]
            if build_numbers_to_add:
                update_data.setdefault('$addToSet', {}).setdefault('metadata.build_numbers', {}).setdefault('$each', []).extend(build_numbers_to_add)

            # Handle 'products' (assuming these are now stored as strings)
            existing_products = set(existing_doc['metadata'].get('products', []))
            new_products = set(cve['metadata'].get('products', []))
            products_to_add = list(new_products - existing_products)
            if products_to_add:
                update_data.setdefault('$addToSet', {}).setdefault('metadata.products', {}).setdefault('$each', []).extend(products_to_add)

            # Perform the update on the current document
            if update_data:
                update_result = db_collection.update_one({'id_': existing_doc['id_']}, {'$addToSet': update_data.get('$addToSet', {}), '$set': update_data.get('$set', {})})

                # Check if the update was successful
                if update_result.matched_count > 0:
                    if update_result.modified_count > 0:
                        print(f"Update successful for document with id_: {existing_doc['id_']}")
                    else:
                        print(f"Document with id_: {existing_doc['id_']} matched but was not modified.")
                else:
                    print(f"No document found with id_: {existing_doc['id_']} to update.")
```

```python
# modify the function that already exists to handle if the first argument isn't passed
def setup_selenium_browser(headless=True):
    """
    Initializes and returns a Selenium WebDriver instance with specified options.
    
    Parameters:
    
    - headless: Boolean indicating whether the browser should run in headless mode.
    
    Returns:
    - driver: A configured Selenium WebDriver instance.
    """
    chrome_options = Options()
    
    if headless:
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--window-size=1920,1080")
    chrome_options.add_argument("--disable-gpu")  # Recommended as per Selenium documentation
    chrome_options.add_argument("--no-sandbox")  # Recommended for running as root/admin

    
    # Create the Chrome WebDriver instance
    driver = webdriver.Chrome(options=chrome_options)
    
    return driver

#  add function to one of the utility files

def parse_package_details(link_text):
    # Modified regular expression pattern
    pattern = re.compile(
        r'(?P<date>\d{4}-\d{2})\s+' # Capture the date (year-month)
        r'(?P<update_type>Dynamic Cumulative|Cumulative)\s+Update\s+for\s+' # Capture the update type
        r'Windows\s+(?P<product_version>\d+)' # Capture the product version (e.g., Windows 10 or Windows 11)
        r'(\s+Version\s+(?P<version>\d{2}H\d))?' # Optionally capture the version (e.g., 21H2), making it optional
        r'\s+for\s+' # Adjusted for potential absence of the version part
        r'(?P<architecture>x64|x86|arm64|ARM64)-based\s+Systems\s+' # Capture the architecture
        r'\(KB(?P<kb_number>\d+)\)' # Capture the KB number
    )

    # Search for matches in the link text
    match = pattern.search(link_text)
    if match:
        # If a match is found, extract the components
        details = match.groupdict()
        # Ensure version key exists, set to 'N/A' if not found
        details['version'] = details.get('version', 'N/A')
        return details
    else:
        # If no match is found, return a dict with default values
        return {
            'date': 'N/A',
            'update_type': 'N/A',
            'product_version': 'N/A',
            'version': 'N/A',
            'architecture': 'N/A',
            'kb_number': 'N/A'
        }
```

#### Function to create a Hash field for each downloadable package

```python
import hashlib

# Connect to MongoDB
mongo_url = f"mongodb+srv://{username}:{password}@bighatcluster.wamzrdr.mongodb.net/"
mongo_client = MongoClient(mongo_url)
db = mongo_client["report_docstore"]
db_update_packages = db["microsoft_update_packages"]

# Fetch documents
cursor = db_update_packages.find({})
documents = list(cursor)
print(len(documents))
for doc in documents:
    # Ensure downloadable_packages field exists and is a list
    if 'downloadable_packages' in doc and isinstance(doc['downloadable_packages'], list):
        updates = []
        for item in doc['downloadable_packages']:
            # Concatenate the unique fields
            unique_string = f"{doc['id']}-{item.get('product_name', '')}-{item.get('product_version', '')}-{item.get('product_architecture', '')}-{item.get('update_type', '')}"
            # Generate a MD5 hash of the concatenated string
            hash_value = hashlib.md5(unique_string.encode()).hexdigest()
            # Prepare the update (add hash to each downloadable package)
            item['hash'] = hash_value
            updates.append(item)
        # print(updates)
        # Update the document in the database with new hashes
        db_update_packages.update_one(
            {'_id': doc['_id']},
            {'$set': {'downloadable_packages': updates}}
        )
        
```
